## b. k-fold lightGBM, k-fold 교차검증 (k-fold model)

- 과대적합 방지
- 모델 간 일반적인 성능 비교
- validation set, development set

- 데이콘 제출 전에 모델 자체 검증가능


```python
from sklearn.model_selection import KFold
```

```python
k_fold = KFold(n_splits = 5, shuffle = True, random_state=777)
# 순서에 의한 영향을 없애기 위해 셔플 진행. 시계열같이 순서에 의미가 있는 경우에는 셔플X
```

```python
for train_idx, val_idx in k_fold.split(X_train):
    print(len(train_idx), len(val_idx))
    break
# 5등분 한 것을 알 수 있다.
```

```python
model = lgb.LGBMRegressor(random_state = 777, n_estimators = 1000)

models = [] # 교차검증을 위한 바구나 5개

for train_idx, val_idx in k_fold.split(X_train):
    x_t = X_train.iloc[train_idx]
    y_t = y_train.iloc[train_idx]
    x_val = X_train.iloc[val_idx]
    y_val = y_train.iloc[val_idx]
    
    models.append(model.fit(x_t, y_t, eval_set = (x_val, y_val), early_stopping_rounds=100, verbose = 100))
    
# early_stopping_rounds=100 100번을 진행해서 오차가 유의미하게 줄어들지 않는다면 멈춰라.
# 그 평가 기준을 5등분한것중 하나인 eval_set으로 검정하는 것
# verbose = 100은 위에서 n_estimators = 1000으로 하여 1000번의 학습을 하는데 그 중 100번마다 출력을 해달라는 것
```

```python
preds = []
for model in models:
    preds.append(model.predict(X_test))
len(preds)
# 5개의 모델들의 예측값을 preds에 저장!
```

```python
kfoldLightGBM = submission.copy() #답안지 하나 복사
```

```python
import numpy as np
```

```python
kfoldLightGBM['box_off_num'] = np.mean(preds, axis = 0) # 예측값들을 평균냄.
# axis = 0 은 축을 0으로하겠다는 것인데 243개 행 각각을 평균내겠다 라는 의미이다.
```

```python
kfoldLightGBM.to_csv('kfoldLightGBM.csv', index = False)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
train['genre'] = le.fit_transform(train['genre']) # 범주형 변수들을 숫자로 다 바꿔준다! 고유번호 하나씩을 부여해줌.
```

```python
train['genre']
```

```python
test['genre'] = le.transform(test['genre']) 
# 위에랑 명령어가 살짝 다른데, fit을 뺀 transform만 씀
# train에서 쓰는 번호 그대로 가져온다.
```

```python
features = ['time', 'dir_prev_num', 'num_staff', 'num_actor', 'dir_prev_bfnum', 'genre']
```

```python
X_train, X_test, y_train = train[features], test[features], train[target]
```

```python
# 값 변경 후 마찬가지로 학습 진행

model = lgb.LGBMRegressor(random_state = 2, n_estimators = 1000)

models = []

for train_idx, val_idx in k_fold.split(X_train):
    x_t = X_train.iloc[train_idx]
    y_t = y_train.iloc[train_idx]
    x_val = X_train.iloc[val_idx]
    y_val = y_train.iloc[val_idx]
    
    models.append(model.fit(x_t, y_t, eval_set = (x_val, y_val), early_stopping_rounds=100, verbose = 100))
```

```python
preds = []
for model in models:
    preds.append(model.predict(X_test))
len(preds)
# 5개의 모델들의 예측값을 preds에 저장!
```

```python
feLightGBM = submission.copy()
```

```python
feLightGBM['box_off_num'] = np.mean(preds, axis=0)
```

```python
feLightGBM.to_csv('feLightGBM.csv', index = False)
```

##  d. grid search (hyperparameter tuning)

튜닝을 해서 성능을 올릴 수 있는지 확인 (모델의 다양한 옵션들 바꿔가며)



```python
from sklearn.model_selection import GridSearchCV
```

```python
model = lgb.LGBMRegressor(random_state=777, n_estimators = 1000)

params = { # 실험해보고자 하는 것들을 넣음.
    'learning_rate':[0.1, 0.01, 0.003], # 부스팅과정에서 얼마나 이전 모델의 영향을 받는지.
    'min_child_samples' : [20, 30]}

gs = GridSearchCV(estimator=model,
                  param_grid=params,
                  scoring = 'neg_mean_squared_error',
                  cv = k_fold)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_params_ # 가장 성능이 좋았던 것
```

```python
# 값 변경 후 마찬가지로 학습 진행

model = lgb.LGBMRegressor(random_state = 777, n_estimators = 1000, learning_rate = 0.003, min_child_samples = 30)

models = []

for train_idx, val_idx in k_fold.split(X_train):
    x_t = X_train.iloc[train_idx]
    y_t = y_train.iloc[train_idx]
    x_val = X_train.iloc[val_idx]
    y_val = y_train.iloc[val_idx]
    
    models.append(model.fit(x_t, y_t, eval_set = (x_val, y_val), early_stopping_rounds=100, verbose = 100))
```

```python
preds = []
for model in models:
    preds.append(model.predict(X_test))
len(preds)
# 5개의 모델들의 예측값을 preds에 저장!
```

```python
gs.best_score_
```

```python
gslgbm = submission.copy()
```

```python
gslgbm['box_off_num'] = np.mean(preds, axis = 0)
```
